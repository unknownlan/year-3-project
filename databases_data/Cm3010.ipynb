## 2. Processing the Raw data

```python

# Load the CSV file
file_path = 'World_population/population.csv'
population_df = pd.read_csv(file_path)

population_df = population_df.dropna(subset=['Code'])
population_df = population_df[population_df['Code'].str.strip() != '']

# Filter the DataFrame to keep only the years from 2014 to 2021
filtered_population_df = population_df[(population_df['Year'] >= 2014) & (population_df['Year'] <= 2021)]

# Save the filtered DataFrame to a new CSV file
filtered_csv_path = 'World_population/filtered_population_2014_2021.csv'
filtered_population_df.to_csv(filtered_csv_path, index=False)

print(f"Filtered data saved to {filtered_csv_path}")
```

### 2.1 only selected years are required and drop the unnecessary years

```python
# Drop columns for years 1960 to 2012
years_to_drop = [str(year) for year in range(1960, 2014)]
df.drop(columns=years_to_drop, inplace=True, errors='ignore')
df.drop(columns='Unnamed: 68', inplace=True, errors='ignore')
df_age_dependency.drop(columns=years_to_drop, inplace=True, errors='ignore')
df_age_dependency_youth.drop(columns=years_to_drop, inplace=True, errors='ignore')
df_age_dependency_elderly.drop(columns=years_to_drop, inplace=True, errors='ignore')

#drop 2023 seperatly
df.drop(columns='2023', inplace=True, errors='ignore')
df_age_dependency.drop(columns='2023', inplace=True, errors='ignore')
df_age_dependency_youth.drop(columns='2023', inplace=True, errors='ignore')
df_age_dependency_elderly.drop(columns='2023', inplace=True, errors='ignore')

# Display the first few rows of the DataFrame to verify the structure
print("First few rows of the DataFrame after cleaning:")
display(df.head())
display(df_age_dependency.head())

# Check the remaining column names
print("\nRemaining column names:")
print(df.columns)
```

### check for missing value

```python
# Clean and Check for missing values
df = df.replace({np.nan: 0})
missing_values = df.isnull().sum()
print("\nMissing values in each column:")
print(missing_values)
```

```python
# Function to process the dataframes
def process_dataframe(data, skiprows):
    # Read the CSV file
    df = data
    
    df = df.replace({np.nan: 0})
    
    # Check for missing values
    missing_values = df.isnull().sum()
    print(missing_values)
    
    # Drop rows with any missing values
    df_cleaned = df.dropna()
    
    # Display the first few rows of the DataFrame to verify the structure
    #print(f"\nFirst few rows of the DataFrame after dropping rows with missing values for {file_path}:")
    display(df_cleaned.head())
    
    return df_cleaned

# Process each dataframe
df_health_cleaned = process_dataframe(df, skiprows=0)
df_working_age_cleaned = process_dataframe(df_age_dependency, skiprows=0)
df_youth_cleaned = process_dataframe(df_age_dependency_youth, skiprows=0)
df_elderly_cleaned = process_dataframe(df_age_dependency_elderly, skiprows=0)

# Check the remaining column names for each DataFrame
print("\nRemaining column names in the health expenditure DataFrame:")
print(df_health_cleaned.columns)

print("\nRemaining column names in the age dependency working age DataFrame:")
print(df_working_age_cleaned.columns)

print("\nRemaining column names in the age dependency youth DataFrame:")
print(df_youth_cleaned.columns)

print("\nRemaining column names in the age dependency elderly DataFrame:")
print(df_elderly_cleaned.columns)
```
